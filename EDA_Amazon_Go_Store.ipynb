```python
#Creating Loss Dataframe
#Now letâ€™s divide the overall data to draw some more accurate conclusions. 
#Create a new Dataframe, where profit is negative which means loss, and concentrate 
#on these areas to improve.
loss_df =amazongostore_df[amazongostore_df['Profit'] < 0]
```

```python
#This will create a new df with all the features where profit is less than Zero.
#Now View it.
loss_df
```

```python
#View the shape of loss df.
loss_df.shape
```

```python
#This loss df contains 1871 rows and 13 columns.
#We can see that there are 9994 rows in the overall superstore Dataframe. 
#Now we have only 1871 rows that are related to loss.
#View the statistical description of the loss Dataframe.
loss_df.describe()
```

```python
Total_loss=np.negative(loss_df['Profit'].sum())
print("Total loss = %.2f" %Total_loss)
```

```python
loss_df.groupby(by='Segment').sum()
```

```python
#More discount leads to more loss, so, to make more profit provide fewer discounts.
loss_df.groupby(by='Prod-Sub-Category').sum()
```

```python
#We can observe more loss in the Binders category, machines category, and tables category 
#when compared to other categories.

#Binders are more getting sold. So even giving less discount may lead to vast loss.

#So better to give discounts on which are getting less sold so that even 
#they will start getting sold more.

loss_df['Prod-Sub-Category'].value_counts()
```

```python
loss_df.groupby(by='City').sum().sort_values('Profit',ascending=True).head(10)
```

```python
#The above-mentioned list shows the names of the top 10 cities where loss is high.
loss_df.sort_values(['Sales'],ascending=True).groupby(by='Prod-Sub-Category').mean()
```

```python
#Here Consumer segment sales might be less when compared to other segments,
#but this is the only segment that provides the highest profits. 
#So, if we increase sales in this Segment by advertisements or something else then, 
#for sure, we can gain more profits.
amazongostore_df.groupby(by='Region').sum()
```
